```python
import pandas as pd
data=pd.read_csv("combined.csv")
pos=[]
negs=[]

for i in range(len(data["sentiment"])):
    if data["sentiment"][i]=="negative":
        negs.append(data["review"][i])
    else:
        pos.append(data["review"][i])
print(len(pos),len(negs))
dataTotrain=pd.DataFrame()
dataTotest=pd.DataFrame()
dataTotrain["review"]=pos[:50000]+negs[:50000]
dataTotrain["sentiment"]=[1 for k in range(50000)]+[0 for k in range(50000)]
dataTotest["review"]=pos[:17500]+negs[:17500]
dataTotest["sentiment"]=[1 for k in range(17500)]+[0 for k in range(17500)]

dataTotrain.to_csv("sentiment_Train.csv")
dataTotest.to_csv("sentiment_Test.csv")



#negtive bais 
```

```python
import nltk
from nltk import word_tokenize

import pandas as pd
import seaborn as sea
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemm=WordNetLemmatizer()


data=pd.read_csv('sentiment_Train.csv')
test=pd.read_csv("sentiment_Test.csv")

##creating words which have influence on sentiments and lemmatizing them

words=set()

emotionalwords=["RB","RBR","RBS","JJ","JJR","JJS","NNP","NN","NNS","VB","VBD","VBG","VBN","VBP"]
for i in range(0,len(data['review']),10):
    jk=word_tokenize(str(data['review'][i]))
    for k in nltk.pos_tag(jk):
        if k[1] in emotionalwords:
            words.add(lemm.lemmatize(k[0].lower()))
        else:
            pass
words=list(words)
len(words)
words[:10]
print(words)
```

```python
emotionalwords=["RB","RBR","RBS","JJ","JJR","JJS","NNP","NN","NNS","VB","VBD","VBG","VBN","VBP"]
def process(GivenText):
    words=[]
    jk=nltk.word_tokenize(GivenText)
    for k in nltk.pos_tag(jk):
        if k[1] in emotionalwords:
            words.append(lemm.lemmatize(k[0].lower()))
        else:
            pass
    return words
def procces2(list12):
    tokens=[0 for i in range(len(words))]
    for jk in list12:
        try:
            tokens[words.index(jk)]+=1
        except:
            pass
        
    
    return tokens
dataOfFeatures=[]
classes=[]

for i in range(0,len(data['review']),5):
    dataOfFeatures.append(procces2(process(str(data['review'][i])))) 
    classes.append([data['sentiment'][i]])

X_test=[]
Y_test=[]
for i in range(0,len(test['review']),2):
    X_test.append(procces2(process(test['review'][i]))) 
    Y_test.append([test['sentiment'][i]])

print(classes[:10])
print(dataOfFeatures[:10])
print(len(classes),len(dataOfFeatures))

print(len(X_test),len(Y_test))


```

```python

import tensorflow as tf
from tensorflow import keras
from keras.callbacks import EarlyStopping
    
# Sample data (replace with your actual data)
X = tf.constant(dataOfFeatures)  # Input data points
y = tf.constant(classes)  # Labels (0 or 1)
X_val=tf.constant(X_test)
y_val=tf.constant(Y_test)


# Define the neural network architecture
model = tf.keras.Sequential([
  tf.keras.layers.Dense(100, activation='relu', input_shape=(len(words),)),  # Hidden layer with 4 neurons and ReLU activation
  tf.keras.layers.Dense(1, activation='sigmoid'),                             # Output layer with 1 neuron and sigmoid activation (for binary classification)
  tf.keras.layers.Dropout(0.5)                   
])

# Compile the model 
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=3)  # Stop after 3 epochs of no validation loss improvement

# Train the model with the callback
model.fit(X,y, epochs=10, validation_data=(X_val,y_val), callbacks=[early_stopping])



# Evaluate the model 
loss, accuracy = model.evaluate(X, y)
print('Test loss:', loss)
print('Test accuracy:', accuracy)


model.save('sentiment_analysis_model.h5')
# Use the model for prediction 
predictions = model.predict(X)





```
